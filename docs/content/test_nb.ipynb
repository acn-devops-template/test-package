## Credit limit columns

```python
credit_limits = spark.sql("""
  SELECT
    rm_key,
    SUM(CASE WHEN row_number = 1 THEN cr_lmt ELSE 0 END) 
      AS dbank_total_valid_crcard_cr_lmt_mthly_v1_0_0_alpha,
    SUM(CASE WHEN row_number = 1 AND UPPER(spplmt_type) NOT IN ('3_REAL', '9_N/A') THEN cr_lmt ELSE 0 END) 
      AS dbank_valid_crcard_primary_cr_lmt_mthly_v1_0_0_alpha,
    SUM(CASE WHEN row_number = 1 AND UPPER(spplmt_type) = '3_REAL' THEN cr_lmt ELSE 0 END) 
      AS dbank_valid_crcard_real_cr_lmt_mthly_v1_0_0_alpha
  FROM (
    SELECT
      rm_key,
      cr_lmt,
      spplmt_type,
      ROW_NUMBER() OVER (PARTITION BY rm_key, asct_card_num ORDER BY spplmt_type, cr_lmt DESC) AS row_number
    FROM dlprod_curated_crcard_db.crcard_card_mthly
    WHERE dl_data_dt = DATE '{}'
    AND LOWER(sts) = 'v'
    AND LOWER(prod_grp) = 'creditcard'
  )
  GROUP BY rm_key
""".format(getArgument('dl_data_dt')))
```

```python
NUM_TOP_MERCHANTS = 3
prefix = 'dbank_valid_crcard_'
suffix = '_mthly_v1_0_0_alpha'

top_merchants = spark.sql("""
  SELECT
    rm_key,
    COLLECT_LIST(
      CASE 
        WHEN mcht_catg IS NOT NULL 
        THEN mcht_catg ELSE 'UNKNOWN' 
      END
    ) AS mcht_catg_list,
    COLLECT_LIST(total_txn) AS total_txn_list
  FROM (
    SELECT
      rm_key,
      mcht_catg,
      COUNT(*) AS total_txn,
      ROW_NUMBER() OVER (
        PARTITION BY rm_key ORDER BY COUNT(*) DESC 
      ) AS row_number
    FROM dlprod_curated_crcard_db.crcard_spc_txn_mthly
    WHERE dl_data_dt = DATE '{0}'
    AND card_num IN (
      SELECT
        card_num
      FROM dlprod_curated_crcard_db.crcard_card_mthly
      WHERE dl_data_dt = DATE '{0}'
      AND LOWER(prod_grp) = 'creditcard'
      AND LOWER(sts) = 'v'
    )
    GROUP BY rm_key, mcht_catg
  )
  WHERE row_number <= {1}
  GROUP BY rm_key
""".format(getArgument('dl_data_dt'), NUM_TOP_MERCHANTS))

top_merchants = top_merchants.select(
  top_merchants.rm_key,
  *[
    sub_elem for i in range(NUM_TOP_MERCHANTS) for sub_elem in (
      top_merchants.mcht_catg_list[i].alias(prefix + f'num_{i+1}_mcht_catg' + suffix),
      top_merchants.total_txn_list[i].alias(prefix + f'num_{i+1}_mcht_txn' + suffix)
    )
  ]
)
```

```python
from pyspark.sql.functions import lit

all_columns = non_credit_limits.columns + credit_limits.columns + top_merchants.columns

final_df = non_credit_limits.join(
  credit_limits,
  on='rm_key',
  how='inner'
).join(
  top_merchants,
  on='rm_key',
  how='left'
).select(
  [non_credit_limits.rm_key] +
  [x for x in all_columns if x != 'rm_key']
)

count_repeated_rm_key_col = 0
for column in all_columns:
  if column == 'rm_key':
    count_repeated_rm_key_col += 1
    
assert len(final_df.columns) == len(all_columns) - count_repeated_rm_key_col + 1

final_df = final_df.withColumn('dl_data_dt', lit(getArgument('dl_data_dt')))
```

## Save results

```python
formatted_save_path = getArgument('save_path') + '_' + getArgument('dl_data_dt').replace('-', '_')

if getArgument('partitionBy') != '':
  final_df.write.partitionBy(getArgument('partitionBy')).format('delta').save(formatted_save_path)
else:
  final_df.write.format('delta').save(formatted_save_path)
```
